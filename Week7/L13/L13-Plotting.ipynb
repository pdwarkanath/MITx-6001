```python
import pylab as plt
```

```python
my_samples = []
my_linear = []
my_squares = []
my_cubes = []
my_exponential = []

for i in range(30):
    my_samples.append(i)
    my_linear.append(i)
    my_squares.append(i**2)
    my_cubes.append(i**3)
    my_exponential.append(1.5**i)
```

```python
plt.plot(my_samples, my_linear)
plt.plot(my_samples, my_squares)
plt.plot(my_samples, my_cubes)
plt.plot(my_samples, my_exponential)
plt.show()
```

```python
plt.figure('lin')
plt.plot(my_samples, my_linear)
plt.figure('square')
plt.plot(my_samples, my_squares)
plt.figure('cube')
plt.plot(my_samples, my_cubes)
plt.figure('expo')
plt.plot(my_samples, my_exponential)
```

```python
plt.figure('lin')
plt.plot(my_samples, my_linear)
plt.xlabel('Sample Points')
plt.ylabel('Linear Function')

plt.figure('square')
plt.plot(my_samples, my_squares)
plt.xlabel('Sample Points')
plt.ylabel('Linear Function')

plt.figure('cube')
plt.plot(my_samples, my_cubes)
plt.xlabel('Sample Points')
plt.ylabel('Cubic Function')

plt.figure('expo')
plt.plot(my_samples, my_exponential)
plt.xlabel('Sample Points')
plt.ylabel('Exponential Function')
```
